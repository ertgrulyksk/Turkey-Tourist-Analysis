```python
finalcountry=countries.groupby(['YEAR']).sum()
finalcountry.head()
```

```python
after=finalcountry[finalcountry.index >= 2015].sum()
after.head()
```

```python
before=finalcountry[finalcountry.index < 2015].sum()
before.head()
```

```python
f=(after/before)-1
flarge=f.nlargest(30)
flarge
```

```python
fsmall=f.nsmallest(30)
fsmall
```

```python
figure(figsize=(16, 8), dpi=120)
plt.barh(fsmall.index,fsmall)
plt.title('Bottom 30 Countries after Calculation ')
plt.ylabel('Country name')
plt.xlabel('Percentage of Tourist Losses')
plt.show()
```

```python
figure(figsize=(16, 8), dpi=120)
plt.barh(flarge.index,flarge, color='blue')
plt.title('Top 30 Countries after Calculation')
plt.ylabel('Country name')
plt.xlabel('Percentage of Tourist Gain')
plt.show()
```
